```python
import torch
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration, pipeline
import gradio as gr
from gtts import gTTS
import IPython.display as ipd

# Load the processor and model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Load the NER pipeline
ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", aggregation_strategy="simple")

# Check if GPU is available and use it
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def generate_caption_with_entities(img):
    """
    Generate a caption for the given image, identify important entities, and include them in the caption.

    Args:
        img (numpy.ndarray): The input image array.

    Returns:
        tuple: The enriched caption and audio file for playback.
    """
    try:
        # Convert the image to a PIL Image
        img_input = Image.fromarray(img)

        # Process the image for the model
        inputs = processor(images=img_input, return_tensors="pt").to(device)

        # Generate a caption
        outputs = model.generate(**inputs)
        caption = processor.decode(outputs[0], skip_special_tokens=True)

        # Identify important entities using NER
        ner_results = ner_pipeline(caption)
        entities = [{"entity": res["entity_group"], "text": res["word"]} for res in ner_results]

        # Enrich the caption by adding entity details
        enriched_caption = caption
        if entities:
            entity_texts = ", ".join([f"{entity['text']} ({entity['entity']})" for entity in entities])
            enriched_caption = f"{caption}. This includes: {entity_texts}."

        # Generate speech from the enriched caption using gTTS
        tts = gTTS(text=enriched_caption, lang="en")
        audio_file = "enriched_caption.mp3"
        tts.save(audio_file)

        # Return the enriched caption and audio file for playback
        return enriched_caption, audio_file
    except Exception as e:
        return f"Error generating caption: {str(e)}", None

# Create a Gradio interface
demo = gr.Interface(
    fn=generate_caption_with_entities,
    inputs=gr.Image(label="Upload an Image", type="numpy"),  # Ensure compatibility with PIL.Image
    outputs=[
        gr.Text(label="Generated Caption with Entities"),
        gr.Audio(label="Audio Caption"),
    ],
    title="Image Captioning with BLIP and Entity Enrichment",
    description="Upload an image, and this app will generate a descriptive caption using the BLIP model, include important persons or monuments in the caption, and read it aloud.",
    examples=[
        ["example_image.jpg"],  # Add some example images for users to test with
    ],
)

# Add a friendly warning if GPU is not available
if device == "cpu":
    print("Warning: GPU not detected. Running on CPU may be slower.")

# Launch the app
demo.launch()
```

```python
#image story
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration, GPT2LMHeadModel, GPT2Tokenizer
import gradio as gr

# Load BLIP image captioning model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Load GPT-2 model and tokenizer for story generation
story_tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
story_model = GPT2LMHeadModel.from_pretrained("gpt2")

# Function to generate caption and story
def generate_story(img):
    img_input = Image.fromarray(img)
    inputs = processor(img_input, return_tensors="pt")
    out = model.generate(**inputs)
    caption = processor.decode(out[0], skip_special_tokens=True)

    # Generate story using GPT-2
    story_prompt = f"The image depicts: {caption} "
    story_inputs = story_tokenizer.encode(story_prompt, return_tensors="pt")
    story_output = story_model.generate(story_inputs, max_length=500, num_return_sequences=1, no_repeat_ngram_size=2)
    story = story_tokenizer.decode(story_output[0], skip_special_tokens=True)

    return story

# Gradio Interface
demo = gr.Interface(
    fn=generate_story,
    inputs=[gr.Image(label="Upload an Image")],
    outputs=[gr.Text(label="Generated Story")],
)

demo.launch()
```

```python
#(final with feedback)

import torch
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
import pytesseract
from gtts import gTTS
import gradio as gr
import hashlib
import json
import os

# Load the processor and model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Check if GPU is available and use it
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# File to store user feedback (image hash to caption mapping)
FEEDBACK_FILE = "feedback_data.json"

# Safely load feedback data if the file exists
try:
    if os.path.exists(FEEDBACK_FILE):
        with open(FEEDBACK_FILE, "r") as f:
            feedback_data = json.load(f)
    else:
        feedback_data = {}
except json.JSONDecodeError:
    feedback_data = {}
    print("Warning: Could not decode feedback_data.json. Starting fresh.")

def hash_image(img):
    """Generate a hash for the image to use as a unique identifier."""
    img_bytes = img.tobytes()
    return hashlib.sha256(img_bytes).hexdigest()

def generate_caption_with_feedback(img):
    """
    Generate a caption for the given image, allowing feedback for correction.

    Args:
        img (numpy.ndarray): The input image array.

    Returns:
        tuple: The generated or retrieved caption, audio for playback, and image hash.
    """
    try:
        # Convert the image to a PIL Image
        img_input = Image.fromarray(img)

        # Hash the image to check for existing feedback
        img_hash = hash_image(img_input)

        # If a corrected caption exists, use it
        if img_hash in feedback_data:
            caption = feedback_data[img_hash]
        else:
            # Generate a descriptive caption
            inputs = processor(images=img_input, return_tensors="pt").to(device)
            outputs = model.generate(**inputs)
            caption = processor.decode(outputs[0], skip_special_tokens=True)

            # Extract text from the image using pytesseract
            detected_text = pytesseract.image_to_string(img_input).strip()
            if detected_text:
                caption += f" Detected text in the image: {detected_text}"

        # Generate speech from the caption using gTTS
        tts = gTTS(text=caption, lang="en")
        audio_file = "caption_feedback.mp3"
        tts.save(audio_file)

        return caption, audio_file, img_hash
    except Exception as e:
        return f"Error generating caption: {str(e)}", None, None

def save_feedback(img_hash, corrected_caption):
    """
    Save corrected captions provided by the user.

    Args:
        img_hash (str): The unique hash for the image.
        corrected_caption (str): The user-provided corrected caption.
    """
    feedback_data[img_hash] = corrected_caption
    with open(FEEDBACK_FILE, "w") as f:
        json.dump(feedback_data, f)

# Create a Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("# SceneScribe: AI Powered Image Captioning")

    with gr.Row():
        img_input = gr.Image(label="Upload an Image", type="numpy")
        caption_output = gr.Textbox(label="Generated Caption")
        audio_output = gr.Audio(label="Audio Caption")

    corrected_caption_input = gr.Textbox(label="Provide Corrected Caption (if incorrect)")
    feedback_button = gr.Button("Submit Corrected Caption")

    feedback_status = gr.Textbox(label="Feedback Status", interactive=False)

    generate_button = gr.Button("Generate Caption")

    def generate_and_display(img):
        caption, audio, img_hash = generate_caption_with_feedback(img)
        return caption, audio, img_hash

    def handle_feedback(img, corrected_caption):
        if img is None:
            return "Error: No image provided."
        if not corrected_caption.strip():
            return "Error: Corrected caption cannot be empty."

        try:
            img_hash = hash_image(Image.fromarray(img))
            save_feedback(img_hash, corrected_caption)
            return f"Corrected caption saved: {corrected_caption}"
        except Exception as e:
            return f"Error saving feedback: {str(e)}"

    generate_button.click(
        generate_and_display,
        inputs=[img_input],
        outputs=[caption_output, audio_output, feedback_status],
    )

    feedback_button.click(
        handle_feedback,
        inputs=[img_input, corrected_caption_input],
        outputs=feedback_status,
    )

# Add a friendly warning if GPU is not available
if device == "cpu":
    print("Warning: GPU not detected. Running on CPU may be slower.")

# Launch the app
demo.launch()
```

```python
import torch
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
import pytesseract
from gtts import gTTS
import gradio as gr
import hashlib
import json
import os
from pathlib import Path
import logging
from datetime import datetime
import numpy as np
from typing import Tuple, Optional, Dict, Any
import threading
import queue
from google.colab import drive
import tempfile
import shutil

# Mount Google Drive to persist feedback data
drive.mount('/content/drive')

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)

class SceneScribe:
    def __init__(self, model_name: str = "Salesforce/blip-image-captioning-base"):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        logger.info(f"Using device: {self.device}")

        # Create directories in Google Drive
        self.drive_dir = Path("/content/drive/MyDrive/SceneScribe")
        self.data_dir = self.drive_dir / "data"
        self.feedback_dir = self.data_dir / "feedback"
        self._create_directories()

        # Create a temporary directory for audio files
        self.temp_dir = tempfile.mkdtemp()
        logger.info(f"Created temporary directory at {self.temp_dir}")

        # Initialize model and processor
        self._init_model(model_name)

        # Load feedback data
        self.feedback_file = self.feedback_dir / "feedback_data.json"
        self.feedback_data = self._load_feedback_data()

    def __del__(self):
        """Cleanup temporary directory when the object is destroyed."""
        try:
            shutil.rmtree(self.temp_dir)
            logger.info(f"Cleaned up temporary directory {self.temp_dir}")
        except Exception as e:
            logger.error(f"Error cleaning up temporary directory: {e}")

    def _create_directories(self):
        """Create necessary directories in Google Drive if they don't exist."""
        for directory in [self.drive_dir, self.data_dir, self.feedback_dir]:
            directory.mkdir(parents=True, exist_ok=True)

    def _init_model(self, model_name: str):
        """Initialize the model and processor with error handling."""
        try:
            self.processor = BlipProcessor.from_pretrained(model_name)
            self.model = BlipForConditionalGeneration.from_pretrained(model_name).to(self.device)
            logger.info("Model loaded successfully")
        except Exception as e:
            logger.error(f"Error loading model: {e}")
            raise

    def _load_feedback_data(self) -> Dict[str, str]:
        """Load feedback data from JSON file in Google Drive."""
        try:
            if self.feedback_file.exists():
                with open(self.feedback_file, "r") as f:
                    return json.load(f)
            return {}
        except json.JSONDecodeError as e:
            logger.error(f"Error loading feedback data: {e}")
            return {}

    @staticmethod
    def hash_image(img: Image.Image) -> str:
        """Generate a secure hash for the image."""
        return hashlib.sha256(np.array(img).tobytes()).hexdigest()

    def generate_caption(self, img: np.ndarray) -> Tuple[str, Optional[str], Optional[str]]:
        """Generate caption for the image with improved error handling and feedback support."""
        try:
            # Convert numpy array to PIL Image
            img_input = Image.fromarray(img)
            img_hash = self.hash_image(img_input)

            # Check for existing feedback
            if img_hash in self.feedback_data:
                caption = self.feedback_data[img_hash]
                logger.info(f"Using feedback caption for image {img_hash[:8]}")
            else:
                # Generate new caption
                inputs = self.processor(images=img_input, return_tensors="pt").to(self.device)
                with torch.no_grad():
                    outputs = self.model.generate(**inputs, max_length=100)
                caption = self.processor.decode(outputs[0], skip_special_tokens=True)

                # Add OCR text if available
                try:
                    ocr_text = pytesseract.image_to_string(img_input).strip()
                    if ocr_text:
                        caption += f" Detected text: {ocr_text}"
                except Exception as e:
                    logger.warning(f"OCR failed: {e}")

            # Generate audio in temporary directory
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            audio_path = os.path.join(self.temp_dir, f"caption_{timestamp}.mp3")

            try:
                tts = gTTS(text=caption, lang="en")
                tts.save(audio_path)
                logger.info(f"Generated audio file at {audio_path}")

                # Read the audio file to verify it exists and is accessible
                with open(audio_path, 'rb') as f:
                    f.read(1024)  # Read first 1KB to verify file

                return caption, audio_path, img_hash
            except Exception as e:
                logger.error(f"Error generating audio: {e}")
                return caption, None, img_hash

        except Exception as e:
            logger.error(f"Error generating caption: {e}")
            return f"Error: {str(e)}", None, None

    def save_feedback(self, img_hash: str, corrected_caption: str) -> str:
        """Save user feedback to Google Drive."""
        try:
            if not corrected_caption.strip():
                return "Error: Caption cannot be empty"

            self.feedback_data[img_hash] = corrected_caption.strip()
            with open(self.feedback_file, "w") as f:
                json.dump(self.feedback_data, f, indent=2)

            logger.info(f"Saved feedback for image {img_hash[:8]}")
            return "Feedback saved successfully to Google Drive"

        except Exception as e:
            logger.error(f"Error saving feedback: {e}")
            return f"Error saving feedback: {str(e)}"

def create_gradio_interface() -> gr.Blocks:
    """Create Gradio interface optimized for Colab."""
    scenescribe = SceneScribe()

    with gr.Blocks() as demo:
        gr.Markdown("""
        # 🎯 SceneScribe: Advanced Image Captioning
        Upload an image to generate a detailed caption with audio narration.
        Provide feedback to improve future captions!
        """)

        with gr.Row():
            with gr.Column(scale=1):
                img_input = gr.Image(
                    label="📸 Upload Image",
                    type="numpy"
                )
                generate_button = gr.Button("🎬 Generate Caption")

            with gr.Column(scale=1):
                caption_output = gr.Textbox(
                    label="📝 Generated Caption",
                    lines=3,
                    interactive=False
                )
                # Modified audio output configuration
                audio_output = gr.Audio(
                    label="🔊 Audio Narration",
                    type="filepath"  # Specify that we're using filepath
                )

        with gr.Row():
            corrected_caption_input = gr.Textbox(
                label="✏️ Provide Corrected Caption (if needed)",
                lines=2,
                placeholder="Enter your correction here..."
            )
            feedback_button = gr.Button("💡 Submit Feedback")

        feedback_status = gr.Textbox(
            label="📢 Status",
            interactive=False
        )

        def generate_and_display(img):
            if img is None:
                return "Please upload an image first.", None, None
            caption, audio_path, img_hash = scenescribe.generate_caption(img)
            return caption, audio_path, img_hash

        def handle_feedback(img, corrected_caption):
            if img is None:
                return "Error: No image provided."
            return scenescribe.save_feedback(
                SceneScribe.hash_image(Image.fromarray(img)),
                corrected_caption
            )

        generate_button.click(
            generate_and_display,
            inputs=[img_input],
            outputs=[caption_output, audio_output, feedback_status]
        )

        feedback_button.click(
            handle_feedback,
            inputs=[img_input, corrected_caption_input],
            outputs=feedback_status
        )

    return demo

# Create and launch the interface
demo = create_gradio_interface()
demo.launch(debug=True, share=True)
```
