```python
#perfect captioning

import torch
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
import gradio as gr

# Load the processor and model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Check if GPU is available and use it
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def generate_caption(img):
    """
    Generate a caption for the given image.

    Args:
        img (numpy.ndarray): The input image array.

    Returns:
        str: The generated caption.
    """
    try:
        # Convert the image to a PIL Image
        img_input = Image.fromarray(img)

        # Process the image for the model
        inputs = processor(images=img_input, return_tensors="pt").to(device)

        # Generate a caption
        outputs = model.generate(**inputs)

        # Decode the caption
        caption = processor.decode(outputs[0], skip_special_tokens=True)
        return caption
    except Exception as e:
        return f"Error generating caption: {str(e)}"

# Create a Gradio interface
demo = gr.Interface(
    fn=generate_caption,
    inputs=gr.Image(label="Upload an Image", type="numpy"),  # Ensure compatibility with PIL.Image
    outputs=gr.Text(label="Generated Caption"),
    title="Image Captioning with BLIP",
    description="Upload an image, and this app will generate a descriptive caption using the BLIP model.",
    examples=[
        ["example_image.jpg"],  # Add some example images for users to test with
    ],
)

# Add a friendly warning if GPU is not available
if device == "cpu":
    print("Warning: GPU not detected. Running on CPU may be slower.")

# Launch the app
demo.launch()
```

```python
import torch
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
import gradio as gr

# Load the processor and model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Check if GPU is available and use it
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def generate_caption(img):
    """
    Generate a caption for the given image.

    Args:
        img (numpy.ndarray): The input image array.

    Returns:
        str: The generated caption or an error message.
    """
    try:
        # Convert the image to a PIL Image
        img_input = Image.fromarray(img)

        # Process the image for the model
        inputs = processor(images=img_input, return_tensors="pt").to(device)

        # Generate a caption
        outputs = model.generate(**inputs, max_length=50, num_beams=5)  # Set parameters for better captions

        # Decode the caption
        caption = processor.decode(outputs[0], skip_special_tokens=True)
        return caption
    except Exception as e:
        return f"Error generating caption: {str(e)}"

# Create a Gradio interface
demo = gr.Interface(
    fn=generate_caption,
    inputs=gr.Image(label="Upload an Image", type="numpy"),
    outputs=gr.Text(label="Generated Caption"),
    title="AI-Powered Image Captioning",
    description=(
        "Upload an image, and this application will generate a descriptive caption using the BLIP model. "
        "Perfect for adding captions to your images!"
    ),
    examples=[
        ["example_image.jpg"],  # Replace with the actual path to example images
    ],
    allow_flagging="never",  # Disable flagging to simplify user experience
)

# Print a message about the device being used
if device == "cpu":
    print("Warning: GPU not detected. Running on CPU may be slower.")

# Launch the app
demo.launch()
```

```python
#aesthetic

import torch
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration, BlipForImageTextRetrieval
import gradio as gr
from torch.utils.data import DataLoader, Dataset
from datasets import load_dataset  # Load or create your dataset

# Load the processor and model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Check if GPU is available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# Define a custom dataset for training
class CustomImageCaptionDataset(Dataset):
    def __init__(self, image_paths, captions, processor):
        self.image_paths = image_paths
        self.captions = captions
        self.processor = processor

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = Image.open(self.image_paths[idx]).convert("RGB")
        caption = self.captions[idx]
        inputs = self.processor(images=image, text=caption, return_tensors="pt", padding=True)
        return {key: val.squeeze(0) for key, val in inputs.items()}

# Training function
def train_model(train_dataset, model, processor, epochs=3, batch_size=8, learning_rate=5e-5):
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate)

    model.train()
    for epoch in range(epochs):
        for batch in train_loader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            pixel_values = batch["pixel_values"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, pixel_values=pixel_values, labels=input_ids)
            loss = outputs.loss

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        print(f"Epoch {epoch + 1} completed. Loss: {loss.item():.4f}")

# Prepare data (example setup)
# Replace with your dataset paths and aesthetic captions (aesthetic captions should be short, e.g., "serenity", "vintage").
image_paths = ["path_to_image1.jpg", "path_to_image2.jpg"]  # Replace with real image paths
captions = ["serenity", "vintage"]  # Aesthetic captions
dataset = CustomImageCaptionDataset(image_paths, captions, processor)

# Train the model
train_model(dataset, model, processor, epochs=3)

# Function to generate aesthetic captions
def generate_aesthetic_caption(img):
    """
    Generate a concise, aesthetic caption for the given image.

    Args:
        img (numpy.ndarray): The input image array.

    Returns:
        str: The generated caption.
    """
    try:
        # Convert the image to a PIL Image
        img_input = Image.fromarray(img)

        # Process the image for the model
        inputs = processor(images=img_input, return_tensors="pt").to(device)

        # Generate a caption
        outputs = model.generate(**inputs, max_length=2, num_beams=5)  # Limit caption length
        caption = processor.decode(outputs[0], skip_special_tokens=True)
        return caption
    except Exception as e:
        return f"Error generating caption: {str(e)}"

# Gradio interface
demo = gr.Interface(
    fn=generate_aesthetic_caption,
    inputs=gr.Image(label="Upload an Image", type="numpy"),
    outputs=gr.Text(label="Aesthetic Caption"),
    title="Aesthetic Image Captioning",
    description=(
        "Upload an image to generate a concise and aesthetic caption. "
        "Perfect for artistic inspiration and creative labeling!"
    ),
    examples=[
        ["example_image.jpg"],  # Replace with paths to aesthetic images
    ],
    allow_flagging="never",
)

# Inform the user if running on CPU
if device == "cpu":
    print("Warning: GPU not detected. Running on CPU may be slower.")

# Launch the app
demo.launch()
```
